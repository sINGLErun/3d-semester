```python
#21. Построить график функции в полярных координатах:
#r(φ) = 2 − 2 sin φ + (sin φ*√|cos φ|)/sin φ+7/3, −π ≤ φ ≤ π

import matplotlib.pyplot as plt
from numpy import *
```

```python
phi = arange(-pi, pi, 0.01)
r = 2 - 2*sin(phi) + (sin(phi)*sqrt(abs(cos(phi))))/(sin(phi)+7/3)

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(231, projection='polar')
ax.plot(phi, r, 'red')
ax.grid(True) #создает сетку (grid - сетка)
```
