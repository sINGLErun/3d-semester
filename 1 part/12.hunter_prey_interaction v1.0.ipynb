```python
#22. Моделирование взаимодействия “хищник-жертва” в модели Лотки-Вольтерры.
#• Задать стартовые значения популяций жертв и хищников x, y и коэффициенты связи:
#    α - коэффициент рождаемости жертв
#    β - коэффициент смертности жертв после охоты хищников
#    γ - коэффициент убыли хищников во время голода
#    δ - коэффициент воспроизводства сытых хищников
#• Решить численно или символьно дифференциальные уравнения
#   { dx/dt = (α − βy)x,
#   { dy/dt = (−γ + δx)y.
#• Решения x(t) и y(t) отобразить графически

import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt

def f(z, t):
    x, y = z
    return [(α - β*y)*x, (-γ + δ*x)*y]
```

```python
#• Задать стартовые значения популяций жертв и хищников и коэффициенты связи

α = 10
β = 2.1
γ = 2
δ = 1

x0y0 = [25, 25] #гран. условия: [жертвы, хищники]

t = np.linspace(0, 5, 101) #последняя переменная - то, что конструирует шаг: (число2-число1)/(число3 - 1) = тому какой будет шаг
```

```python
#• Решить численно или(!) символьно дифференциальные уравнения

[x, y] = odeint(f, x0y0, t, full_output = False).T
```

```python
#• Решения x(t) и y(t) отобразить графически

fig = plt.figure(facecolor = 'white')

plt.plot(t, x, 'green', linewidth = 1)
plt.plot(t, y, 'red', linewidth = 1)
plt.xlabel('time')

plt.grid(True)
```
