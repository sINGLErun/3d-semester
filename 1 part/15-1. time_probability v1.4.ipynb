```python
#15. Найти вероятность того, что на цифровых часах с секундами будут одновременно отображаться и “16”, и “37”.
    #сделаем программу, которая по секундам будет выводить время в дне, как на цифровых часах
    #преобразуем эту балду в лист и посчитаем там количество "16" и "37"

from datetime import timedelta
from itertools import groupby
```

```python
#тут есть только одно но - не учитывается время вида 16:16:16 и такого типа
#прикол в том, чтобы такое время появлялось только один раз
#лазейка через set

good_times = [str(timedelta(seconds = i)) for i in range(1, 24*60*60 + 1) for j in range(0,3) if str(timedelta(seconds = i)).split(':')[j] == '16' or str(timedelta(seconds = i)).split(':')[j] == '37']

better_times = [x for x in good_times if '37' in x and '16' in x]
better_times
#len(better_times)
```

```python
#good_times = [el for el, _ in groupby(good_times)]
#good_times
better_times = [el for el, _ in groupby(better_times)]
better_times
#good_times #видно что порядок сильно нарушается от сета
```

```python
#probability = len(good_times)/(24*60*60)
probability = len(better_times)/(24*60*60)
print(probability) #вероятность может оказаться больше из-за того, что последовательные секунды минуты остаются 16, но моё мнение - более правильно
```
