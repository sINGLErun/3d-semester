```python
#16. Вычислить π методом Монте-Карло.
#• Разыграть много случайных x и y в квадрате [-1, 1] x [-1, 1]
#• Подсчитать долю попавших в круг с радиусом 1.
#• По этой доле оценить значение π.

import numpy as np
import matplotlib.pyplot as plt
```

```python
#• Разыграть много случайных x и y в квадрате [-1, 1] x [-1, 1]

dot_coords = (-2)*np.random.sample((100000,2)) + 1
dot_in_circle = [z for z in dot_coords if z[0]**2 + z[1]**2 < 1]

#print(dot_coords)
print(f'\nThere is {len(dot_coords)} dots and {len(dot_in_circle)} of them in the unit circle')
```

```python
#• Подсчитать долю попавших в круг с радиусом 1.

P = len(dot_in_circle)/len(dot_coords)
print(P)
pi = P*4
print(f'pi in this round is: ({pi}, {np.pi - pi})')
```

```python
fig = plt.figure(figsize = (5, 5))

dot_not_in_circle_x = set([z[0] for z in dot_coords]) - set([z[0] for z in dot_in_circle]) #с двумерными массивами нельзя обращаться как с одномерными множествами
#но что если все пары чисел перевести в комплексные числа, их сравнить как множества, а потом вернуть все
dot_not_in_circle = [list(z) for z in dot_coords if z[0] in dot_not_in_circle_x]

phi = np.arange(0, 2*np.pi, 0.01)
plt.plot(np.cos(phi), np.sin(phi), c = 'purple')  

for z in dot_not_in_circle:
    plt.scatter(z[0], z[1], linewidths = 0)
for z in dot_in_circle:
    plt.scatter(z[0], z[1], c = 'orange', linewidths = 0)
```
